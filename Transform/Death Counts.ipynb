```python
#import dependencies
import os
import csv
import pandas as pd
```

```python
# Path to collect data from the Resources folder

#1 - Death Counts for Infections - Influenza, Pnuemonia, COVID-19

file_path = os.path.join('..', 'Resources', 'Death_Counts_for_infections_Influenza__Pneumonia__and_COVID-19.csv')

# Import the Death_Counts_for_infections_Influenza__Pneumonia__and_COVID-19.csv file as a DataFrame
death_counts_df=pd.read_csv(file_path)

death_counts_df.head(5)
```

```python
# Finding how many unique Jurisdictions are provided
death_counts_df["Jurisdiction"].unique() ##value_counts()

####  HHS Regions: https://www.hhs.gov/about/agencies/iea/regional-offices/index.html
### exclude since we cannot aggregate with other datasources
```

```python
##remove HHS Regions from data
clean_df = death_counts_df[~death_counts_df["Jurisdiction"].str.contains('HHS Region')] 
clean_df["Jurisdiction"].unique()
```

```python
clean_df.describe()
```

```python
clean_df2=clean_df[clean_df["Age Group"]!="All Ages"]
clean_df2
```

```python
#Group by Jurisdiction and Age Group - mean of remaining columns
age_grp1_df=clean_df2.groupby(['Jurisdiction', 'Age Group']).mean()
age_grp1_df
age_grp1_df=age_grp1_df.rename(
    columns= {'COVID-19 Deaths':"Average COVID-19 Deaths",
              'Total Deaths':"Average Total Deaths",
              "Pneumonia Deaths":"Average Pneumonia Deaths",
              "Influenza Deaths":"Average Influenza Deaths",
              "Pneumonia or Influenza":"Average Pneumonia or Influenza",
              "Pneumonia, Influenza, or COVID-19 Deaths":"Average Pneumonia, Influenza, or COVID-19 Deaths"         
            })
age_grp1_df.head(5)




```

```python
#Group by Jurisdiction and Age Group - sum of remaining columns
age_grp2_df=clean_df2.groupby(['Jurisdiction', 'Age Group']).sum()
age_grp2_df.head(5)
```

```python
#Group by Jurisdiction  - mean of remaining columns
age_grp3_df=clean_df2.groupby(['Jurisdiction']).mean()
age_grp3_df=age_grp3_df.rename(
    columns= {'COVID-19 Deaths':"Average COVID-19 Deaths",
              'Total Deaths':"Average Total Deaths",
              "Pneumonia Deaths":"Average Pneumonia Deaths",
              "Influenza Deaths":"Average Influenza Deaths",
              "Pneumonia or Influenza":"Average Pneumonia or Influenza",
              "Pneumonia, Influenza, or COVID-19 Deaths":"Average Pneumonia, Influenza, or COVID-19 Deaths"         
            })
age_grp3_df.head(5)

```
