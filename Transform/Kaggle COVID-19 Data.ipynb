```python
#import dependencies
import os
import csv
import pandas as pd
```

```python
# Path to collect data from the Resources folder

#1 - COVID-19 - Confirmed, Deaths and Recovered Counts

file_path = os.path.join('..', 'Resources', 'Kaggle_Covid_19_worldwide_data_set.csv')

# Import the Kaggle_Covid_19_worldwide_data_set.csv file as a DataFrame
covid_df=pd.read_csv(file_path)

covid_df.head(5)
```

```python
####Cleansing

covid_df['Country/Region'].unique()
```

```python
covid_df['Country/Region']= covid_df['Country/Region'].replace({
    "('St. Martin',)"  : 'St. Martin'
    })
covid_df
```

```python
covid_df['ObservationDate_dt']=pd.to_datetime(covid_df['ObservationDate'], format='%m/%d/%Y')
```

```python
agg_dict={
    'Confirmed': [pd.Series.sum, pd.Series.mean],
    'Deaths': [pd.Series.sum, pd.Series.mean],
    'Recovered': [pd.Series.sum, pd.Series.mean]
}
covid_df_grouped=covid_df.set_index('ObservationDate_dt')

covid_df_grouped=covid_df_grouped.groupby([pd.Grouper(freq='M'), 'Country/Region', 'Province/State']).\
                          agg(agg_dict)
```

```python
covid_df_grouped.reset_index(inplace=True)
```
