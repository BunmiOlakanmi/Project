```python
#import dependencies
import os
import csv
import pandas as pd
```

```python
# Path to collect data from the Resources folder

#1 - Unemployment rates over the period Mar2020 - Sep 2020

file_path = os.path.join('..', 'Resources', 'Unemployment_rate_worldwide_Mar20 to Sept20.csv')

# Import the Unemployment_Rates_Canada_only.csv file as a DataFrame
unemployment_rate_df=pd.read_csv(file_path)

unemployment_rate_df.head(5)
```

```python
unemployment_rate_df['LOCATION'].unique()
```

```python
##remove from data"LOCATION" : OECD, G-7, EA19, EU27_2020

clean_unemp_df = unemployment_rate_df[~unemployment_rate_df['LOCATION'].str.contains('OECD')] 
clean_unemp_df = clean_unemp_df[~clean_unemp_df['LOCATION'].str.contains('G-7')] 
clean_unemp_df = clean_unemp_df [~clean_unemp_df['LOCATION'].str.contains('EA19')] 
clean_unemp_df = clean_unemp_df [~clean_unemp_df['LOCATION'].str.contains('EU27_2020')] 

clean_unemp_df["LOCATION"].unique()
```

```python
###Convert Country Codes to Country Name

country_code_df= pd.read_html('https://www.iban.com/country-codes')
country_code_df[0]

clean_code_df= country_code_df[0]
clean_code_df=clean_code_df.rename(columns = {'Alpha-3 code' : "LOCATION"})
clean_code_df
```

```python
merge_df = pd.merge(clean_unemp_df, clean_code_df, on="LOCATION", how="left")

merge_df
```
